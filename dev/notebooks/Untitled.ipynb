```python
import pylapsy as ply
```

```python
images = ply.io.get_testimg_files_deshake()
images
```

```python
first_file, second_file = images[0], images[1]
```

```python
first = ply.Image(first_file)
first.show();
```

```python
second = ply.Image(second_file)
second_gray = second.to_gray()
```

```python
first_gray = first.to_gray()
first_gray.show()
```

```python
points_to_track_first = ply.utils.find_good_features_to_track(first_gray.img)
points_to_track_second = ply.utils.find_good_features_to_track(second_gray.img)
```

```python
ax = first.show()
ply.utils.plot_feature_points(points_to_track_first, ax=ax, color='r')
ply.utils.plot_feature_points(points_to_track_second, ax=ax, color='lime')
```
